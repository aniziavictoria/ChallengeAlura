### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
#faturamento total por loja
faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

#Resultados
print(f'Faturamento Loja 1: R${faturamento_loja1:.2f}')
print(f'Faturamento Loja 2: R${faturamento_loja2:.2f}')
print(f'Faturamento Loja 3: R${faturamento_loja3:.2f}')
print(f'Faturamento Loja 4: R${faturamento_loja4:.2f}')
```

# 2. Vendas por Categoria


```python
#quantidade de produtos vendidos por categoria em cada loja
produtos_loja1 = loja['Categoria do Produto'].value_counts()
produtos_loja2 = loja2['Categoria do Produto'].value_counts()
produtos_loja3 = loja3['Categoria do Produto'].value_counts()
produtos_loja4 = loja4['Categoria do Produto'].value_counts()
#resultados
print(f'Produtos vendidos na loja 1: \n{produtos_loja1}')
print(f'Produtos vendidos na loja 2: \n{produtos_loja2}')
print(f'Produtos vendidos na loja 3: \n{produtos_loja3}')
print(f'Produtos vendidos na loja 4: \n{produtos_loja4}')
```

# 3. Média de Avaliação das Lojas

```python
avaliacao_media_loja1 = loja['Avaliação da compra'].mean()
avaliacao_media_loja2 = loja2['Avaliação da compra'].mean()
avaliacao_media_loja3 = loja3['Avaliação da compra'].mean()
avaliacao_media_loja4 = loja4['Avaliação da compra'].mean()
print (f'Media de Avaliação - Loja 1: {avaliacao_media_loja1:.2f}')
print (f'Media de Avaliação - Loja 2: {avaliacao_media_loja2:.2f}')
print (f'Media de Avaliação - Loja 3: {avaliacao_media_loja3:.2f}')
print (f'Media de Avaliação - Loja 4: {avaliacao_media_loja4:.2f}')
```

# 4. Produtos Mais e Menos Vendidos

```python
# Produtos mais e menos vendidos por loja

# Loja 1
mais_vendidos1 = loja['Produto'].value_counts().head(5)
menos_vendidos1 = loja['Produto'].value_counts().tail(5)

# Loja 2
mais_vendidos2 = loja2['Produto'].value_counts().head(5)
menos_vendidos2 = loja2['Produto'].value_counts().tail(5)

# Loja 3
mais_vendidos3 = loja3['Produto'].value_counts().head(5)
menos_vendidos3 = loja3['Produto'].value_counts().tail(5)

# Loja 4
mais_vendidos4 = loja4['Produto'].value_counts().head(5)
menos_vendidos4 = loja4['Produto'].value_counts().tail(5)

# Exibir os resultados
print("Loja 1 - Mais vendidos:\n", mais_vendidos1)
print("\nLoja 1 - Menos vendidos:\n", menos_vendidos1)

print("\nLoja 2 - Mais vendidos:\n", mais_vendidos2)
print("\nLoja 2 - Menos vendidos:\n", menos_vendidos2)

print("\nLoja 3 - Mais vendidos:\n", mais_vendidos3)
print("\nLoja 3 - Menos vendidos:\n", menos_vendidos3)

print("\nLoja 4 - Mais vendidos:\n", mais_vendidos4)
print("\nLoja 4 - Menos vendidos:\n", menos_vendidos4)
```

# 5. Frete Médio por Loja


```python
# Custo médio de frete por loja
frete_medio_loja1 = loja['Frete'].mean()
frete_medio_loja2 = loja2['Frete'].mean()
frete_medio_loja3 = loja3['Frete'].mean()
frete_medio_loja4 = loja4['Frete'].mean()

# Resultados
print(f'Custo médio de frete - Loja 1: R${frete_medio_loja1:.2f}')
print(f'Custo médio de frete - Loja 2: R${frete_medio_loja2:.2f}')
print(f'Custo médio de frete - Loja 3: R${frete_medio_loja3:.2f}')
print(f'Custo médio de frete - Loja 4: R${frete_medio_loja4:.2f}')
```

```python
import matplotlib.pyplot as plt

# --- Gráfico 1: Faturamento total por loja ---
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
faturamentos = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]

plt.figure(figsize=(8,5))
plt.bar(lojas, faturamentos, color='skyblue')
plt.title('Faturamento Total por Loja')
plt.ylabel('Faturamento (R$)')
plt.xlabel('Lojas')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# --- Gráfico 2: Distribuição das categorias vendidas na Loja 1 ---
plt.figure(figsize=(7,7))
plt.pie(produtos_loja1.values, labels=produtos_loja1.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribuição de Produtos Vendidos por Categoria - Loja 1')
plt.show()

# --- Gráfico 3: Média das avaliações por loja ---
media_avaliacao_lojas = [avaliacao_media_loja1, avaliacao_media_loja2, avaliacao_media_loja3, avaliacao_media_loja4]
plt.figure(figsize=(8,5))
plt.plot(lojas, media_avaliacao_lojas, marker='o', linestyle='-', color='green')
plt.title('Média das Avaliações por Loja')
plt.ylabel('Avaliação Média')
plt.ylim(3.5, 4.5)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

# Relatório Final

## Introdução

Este relatório tem como objetivo analisar o desempenho das quatro lojas pertencentes ao Senhor João, considerando diferentes aspectos importantes para a tomada de decisão sobre qual loja deve ser vendida. Foram avaliados o faturamento total, a distribuição das categorias de produtos vendidos, a satisfação dos clientes por meio das avaliações, os produtos mais e menos vendidos, e o custo médio de frete de cada loja. A partir dessas análises, será apresentada uma recomendação fundamentada para a melhor decisão.

## Desenvolvimento

### Faturamento Total

O faturamento total é um dos indicadores financeiros mais importantes para avaliar o sucesso de cada loja. Os valores apurados foram:

- Loja 1: R$ 1.534.509,12  
- Loja 2: R$ 1.488.459,06  
- Loja 3: R$ 1.464.025,03  
- Loja 4: R$ 1.384.497,58  

A Loja 1 apresenta o maior faturamento, seguida pela Loja 2, Loja 3 e Loja 4.

### Distribuição das Categorias de Produtos Vendidos

Ao analisar as quantidades de produtos vendidos por categoria, notamos que as categorias mais populares nas quatro lojas são móveis, eletrônicos e brinquedos, com pequenas variações nas posições entre as lojas. As categorias menos populares tendem a ser livros, utilidades domésticas e instrumentos musicais, embora também variem um pouco.

Isso indica que o mix de produtos vendido é relativamente similar entre as lojas, com destaque para móveis e eletrônicos como os principais motores das vendas.

### Média das Avaliações dos Clientes

A satisfação do cliente, medida pela média das avaliações, foi:

- Loja 1: 3,98  
- Loja 2: 4,04  
- Loja 3: 4,05  
- Loja 4: 4,00  

A Loja 3 tem a melhor média de avaliação, seguida de perto pela Loja 2, com a Loja 1 apresentando a menor média entre elas. Isso pode indicar um nível ligeiramente menor de satisfação do cliente na Loja 1, apesar do seu faturamento mais alto.

### Produtos Mais e Menos Vendidos

Cada loja apresenta um conjunto distinto de produtos mais vendidos e menos vendidos, mostrando que há particularidades locais ou de estoque:

- Produtos mais vendidos incluem itens como "Panela de pressão" na Loja 1, "Iniciando em programação" na Loja 2, "Kit banquetas" na Loja 3 e "Cama box" na Loja 4.
- Produtos menos vendidos variam entre "Fone de ouvido", "Poltrona", "Guitarra" e "Geladeira" dependendo da loja.

Essas diferenças devem ser consideradas no contexto da demanda local e das estratégias de venda específicas.

### Custo Médio de Frete

O custo médio de frete impacta diretamente a margem de lucro e a competitividade de preço. Os custos médios foram:

- Loja 1:  34,69 reais
- Loja 2:  33,62 reais
- Loja 3:  33,07 reais
- Loja 4:  31,28 reais

A Loja 4 apresenta o menor custo médio de frete, o que pode ser uma vantagem competitiva, enquanto a Loja 1 tem o maior custo.

## Conclusão e Recomendação

Considerando todos os fatores analisados, a recomendação é que o Senhor João **venda a Loja 4**.

**Justificativas:**

- Apesar de ter o menor faturamento total, a Loja 4 apresenta um custo médio de frete mais baixo, o que pode indicar eficiência logística e menor custo operacional.
- A satisfação dos clientes está em um nível competitivo (média de avaliação 4,00), próxima às outras lojas.
- O mix de produtos vendidos mostra boa diversidade e não apresenta grandes pontos fracos em categorias importantes.
- Produtos mais vendidos na Loja 4 indicam potencial de mercado saudável e alinhado com as preferências locais.

Vender a Loja 4 pode liberar recursos para focar nas lojas que apresentam maior faturamento (Loja 1 e 2) e melhor avaliação, otimizando o portfólio de negócios do Senhor João.
